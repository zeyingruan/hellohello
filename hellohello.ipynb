```python
import pandas as pd
import numpy as np

#read catalog  file
catalogs = pd.read_csv('ttt.csv').values

#columns for products
Seller_Product_Id=[]
Product_Identifier=[]
Selling_Price=[]
Stock=[]
SKU=[]
Quantity_Increment=[]
Min_Order_Quantity=[]
Sustack_Stock=[]
Track_Inventory=[]
Threshold_Stock=[]
Condition_Identifier=[]
Title=[]
Comments=[]
URL_Keyword=[]
Available_From=[]
Active=[]
Cod_Available=[]

#definition to generate products for one catalog
def generate(catalog):
    #read data from each row
    productId = catalog[0]
    Username = catalog[1]
    Name= catalog[2]
    Description= catalog[3]
    Youtube_Video= catalog[4]
    Category_Identifier= catalog[5]
    Brand_Identifier= catalog[6]
    Product_Type_Identifier= catalog[7]
    Model= catalog[8]
    Tax_Category_Identifier= catalog[9]
    Length= catalog[10]
    Width= catalog[11]
    Height= catalog[12]
    Dimension_Unit_Identifier= catalog[13]
    Weight= catalog[14]
    Weight_Unit_Identifier= catalog[15]
    Shipping_Country_Code= catalog[16]
    Free_Shipping= catalog[17]
    Cod_Available_= catalog[18]
    Approved_= catalog[19]
    Active_= catalog[20]
    options = catalog[21].upper()
    option_value = catalog[22].upper()
    same_price = catalog[23].upper()
    price = catalog[24]
    miniquan = catalog[25]
    skunumber = catalog[26].upper()
    date = catalog[27]
    
    #first case: there are different options for one product
    if options == 'YES':
        option_value = option_value.split(',')
        if same_price == 'NO':
            price = price.split(',')
            miniquan = miniquan.split(',')
        id = 1
        for i in option_value:
            Seller_Product_Id.append(str(id))
            Product_Identifier.append(productId)
            #same price for each option value
            if same_price == 'YES':
                Selling_Price.append(price)
                Stock.append(int(miniquan)*10)
                Quantity_Increment.append(miniquan)
                Min_Order_Quantity.append(miniquan)
                Threshold_Stock.append(int(miniquan)*2)
            #different prices for each option value
            else:
                Selling_Price.append(int(price[id-1]))
                Stock.append(int(miniquan[id-1])*10)
                Quantity_Increment.append(int(miniquan[id-1]))
                Min_Order_Quantity.append(int(miniquan[id-1]))
                Threshold_Stock.append(int(miniquan[id-1])*2)
            SKU.append('BW'+skunumber+i.upper())
            Sustack_Stock.append('1')
            Track_Inventory.append('1')
            Condition_Identifier.append('New')
            if Free_Shipping == 'YES':
                Title.append(Name+' - '+i.upper()+' | BW'+skunumber+i.upper()+' | Free Shipping | Wholesale | '+Brand_Identifier)
                URL_Keyword.append(Name+' '+i.upper()+' BW'+skunumber+i.upper()+' Free Shipping Wholesale '+Brand_Identifier)
            else:
                Title.append(Name+' - '+i.upper()+' | BW'+skunumber+i.upper()+' | Wholesale | '+Brand_Identifier)
                URL_Keyword.append(Name+' '+i.upper()+' BW'+skunumber+i.upper()+' Wholesale '+Brand_Identifier)
            Comments.append('')
            Available_From.append(date)
            Active.append(Active_)
            Cod_Available.append(Cod_Available_)
            id+=1
    #second case: no options
    else:
        Seller_Product_Id.append(1)
        Product_Identifier.append(productId)
        Selling_Price.append(price)
        Stock.append(int(miniquan)*10)
        Quantity_Increment.append(miniquan)
        Min_Order_Quantity.append(miniquan)
        Threshold_Stock.append(int(miniquan)*2)
        SKU.append('BW'+skunumber)
        Sustack_Stock.append('1')
        Track_Inventory.append('1')
        Condition_Identifier.append('New')
        if Free_Shipping == 'YES':
            Title.append(Name+' - '+' | BW'+skunumber+' | Free Shipping | Wholesale | '+Brand_Identifier)
            URL_Keyword.append(Name+' '+' BW'+skunumber+' Free Shipping Wholesale '+Brand_Identifier)
        else:
            Title.append(Name+' - '+' | BW'+skunumber+' | Wholesale | '+Brand_Identifier)
            URL_Keyword.append(Name+' '+' BW'+skunumber+' Wholesale '+Brand_Identifier)
        Comments.append('')
        Available_From.append(date)
        Active.append(Active_)
        Cod_Available.append(Cod_Available_)
 

#generate products for all the catalogs
for i in catalogs:
    generate(i)

#column names for product csv file
col = pd.Index(['Seller Product Id','Product Identifier','Selling Price','Stock','SKU','Quantity Increment','Min Order Quantity',
      'Sustack Stock','Track Inventory','Threshold Stock Level','Condition Identifier','Title','Comments','URL Keyword',
   'Available From','Active','Cod Available'],dtype = 'object')

mydict = {'Seller Product Id':Seller_Product_Id,'Product Identifier':Product_Identifier ,'Selling Price':Selling_Price,'Stock':Stock,
          'SKU':SKU,'Quantity Increment':Quantity_Increment,'Min Order Quantity':Min_Order_Quantity,
      'Sustack Stock':Sustack_Stock,'Track Inventory':Track_Inventory,'Threshold Stock Level':Threshold_Stock,'Condition Identifier':Condition_Identifier,'Title':Title,
          'Comments':Comments,'URL Keyword':URL_Keyword, 'Available From':Available_From,'Active':Active,'Cod Available':Cod_Available}

df = pd.DataFrame(mydict,columns = col)
display(df)
     
df.to_csv("testproduct.csv",encoding='utf-8',index = False)
```
